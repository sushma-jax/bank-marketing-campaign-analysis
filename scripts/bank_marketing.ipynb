```python
# Replace special values
if 'pdays' in df.columns:
    df['pdays'] = df['pdays'].replace(999, pd.NA)

# Clean categorical values
cat_cols = df.select_dtypes('object').columns
for col in cat_cols:
    df[col] = df[col].str.lower().str.strip()
```

```python
df.head(20)
```

```python
df['deposit'].value_counts(normalize=True).plot(
    kind='bar', color=['skyblue','salmon'])
plt.title("Target distribution (deposited vs Not)")
plt.ylabel("Proportion")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df['age'], bins=30, kde=True, color='teal')
plt.axvline(df['age'].median(), color='red', linestyle='--', label='Median')
plt.title("Age Distribution")
plt.legend()
plt.show()
```

```python
sns.histplot(df['balance'], bins=40, color="purple")
plt.xlim(-2000, 10000)
plt.title("Customer Balance Distribution")
plt.show()

```

```python
sns.countplot(y='job', data=df, order=df['job'].value_counts().index)
plt.title("Distribution of Jobs")
plt.show()
```

```python
conv = df.groupby('job')['deposit'].apply(lambda x: (x=='yes').mean()).sort_values(ascending=False)

sns.barplot(x=conv.values, y=conv.index, palette="viridis")
plt.title("Conversion Rate by Job")
plt.xlabel("Conversion Rate")
plt.show()

```

```python
conv = df.groupby('contact')['deposit'].apply(lambda x: (x=='yes').mean())
sns.barplot(x=conv.index, y=conv.values, palette="mako")
plt.title("Conversion by Contact Method")
plt.ylabel("Conversion Rate")
plt.show()
```

```python
conv = df.groupby('month')['deposit'].apply(lambda x: (x=='yes').mean())
sns.lineplot(x=conv.index, y=conv.values, marker="o")
plt.title("Conversion Rate by Month")
plt.ylabel("Conversion Rate")
plt.show()
```

```python
num_cols = ['age','balance','duration','campaign','pdays','previous']
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
# Age groups
df['age_group'] = pd.cut(df['age'], bins=[17,30,45,60,100], labels=['18-30','31-45','46-60','60+'])

# Campaign buckets
df['campaign_bucket'] = pd.cut(df['campaign'], bins=[0,1,3,10,100], labels=['1','2-3','4-10','11+'])

# Re-check conversion
sns.barplot(x='campaign_bucket', y=df['deposit'].eq('yes'), data=df, palette="Set2")
plt.title("Conversion by campaign_bucket")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='job', hue='deposit', palette='Set2')
plt.title('Job Type vs. Deposit Subscription', fontsize=14, weight='bold')
plt.xticks(rotation=45)
plt.xlabel('Job Type')
plt.ylabel('Count')
plt.legend(title='Subscribed')
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='education', hue='deposit', palette='coolwarm')
plt.title('Education Level vs. Deposit Subscription', fontsize=14, weight='bold')
plt.xticks(rotation=45)
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.legend(title='Subscribed')
plt.show()
```

```python
# Create age groups (bins)
bins = [18, 25, 35, 45, 55, 65, 100]
labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '65+']
df['Age_Group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Calculate deposit rate for each age group
age_group_rate = df.groupby('Age_Group')['deposit'].value_counts(normalize=True).unstack().fillna(0) * 100

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x=age_group_rate.index, y=age_group_rate['yes'], palette='viridis')
plt.title('Age Group vs. Deposit Rate', fontsize=14, weight='bold')
plt.ylabel('Deposit Subscription Rate (%)')
plt.xlabel('Age Group')
plt.show()
```
